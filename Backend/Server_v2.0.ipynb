# Backend: based on Flask

```python
from flask import Flask, render_template, request, redirect, url_for
import pickle
import numpy as np
```

```python
# Loading trained Models
chronic_dis_pred = pickle.load(open("models/trained_model_Kidney_disease.pkl", "rb"))
diabetes_pred = pickle.load(open("models/trained_model_diabetes_pred_knn.pkl", "rb"))
heartAttackRisk_pred = pickle.load(open("models/trained_model_heartAttack_risk_RFC.pkl", "rb"))
```

```python
app = Flask(__name__)


@app.route("/")
def hello():
    #return "Hello"
    return render_template("/index.html")


@app.route("/chronic_check", methods=['POST'])
def select1():
    return render_template("chronic_form.html")

@app.route("/sugar_check",methods=['POST'])
def select2():
    return render_template("sugar_form.html")

@app.route("/cardiac_check", methods=['POST'])
def select3():
    return render_template("heart_form.html")




@app.route("/chronic", methods=['POST'])
def chronic():
    print(x for x in request.form.values())
    float_features = [float(x) for x in request.form.values()]
    features = [np.array(float_features)]
    print("Features: ", features)
    prediction = chronic_dis_pred.predict(features)
    print("Predictions: ", prediction[0])
    output = prediction[0]
    
    #python to html
    if output==1:
        txt = "Patient have the symptoms and signs of Chronic Disease"
        op_file = "chronic_pos.html"
    else:
        txt = "No traces of Chronic kidney disease found "
        op_file = "chronic_neg.html"
    
    return render_template(op_file,txt = txt)


@app.route("/diabetes", methods=['POST'])
def diabetes():
    float_features = [float(x) for x in request.form.values()]
    features = [np.array(float_features)]
    print("Features: ", features)
    prediction = diabetes_pred.predict(features)
    print("Predictions: ", prediction[0])
    output = prediction[0]
    
    #python to html
    if output==1:
        txt = "Patient might have or is in the High Risk of having Diabetes."
        op_file = "sugar_pos.html"
    else:
        txt = "Patient is out of danger of Diabetes "
        op_file = "sugar_neg.html"
    
    return render_template(op_file,txt = txt)


@app.route("/cardiac", methods=['POST'])
def cardiac_failure():
    float_features = [float(x) for x in request.form.values()]
    features = [np.array(float_features)]
    print("Features: ", features)
    prediction = heartAttackRisk_pred.predict(features)
    print("Prediction: ", prediction[0])
    output = prediction[0]
    
    if output==1:
        txt = "Patient is at High Risk of Heart Failure"
        op_file = "heart_pos.html"
    else:
        txt = 'Low Risk, yet to be cautious'
        op_file = "heart_neg.html"
    #python to html
    return render_template(op_file, txt = txt)





if __name__ == "__main__":
	app.run()
```
